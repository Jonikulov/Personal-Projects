```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/sales_orders_2014.csv")
df
```

## 1. Add a new column `quantity_10`: quantity * 10

```python
df['quantity_10'] = df['quantity'] * 10
df.head()
```

## 2. Add a new column a particular place

```python
# Adding `new_column` after the `region` column
df.insert(loc=6, column='new_column', value=df['discount']*100)
df.head()
```

## 3. Delete a column

```python
df.drop(columns=['new_column', 'quantity_10'], inplace=True)
df.head()
```

## 4. Sorting a DataFrame

```python
df['order_date'] = pd.to_datetime(df['order_date'],
                                  format='%m-%d-%Y',
                                  infer_datetime_format=True)

df.sort_values(by=['order_date', 'region'])
```

## 5. Filter DataFrame (single column / condition)

```python
df[df['country']=='Uzbekistan'].head()
```

## 6. Filter DataFrame (multiple columns / conditions) - AND operator

```python
filter_df = (df['segment']=='Home Office') & (df['category']=='Technology')
df[filter_df]
```

## 7. Filter DataFrame (multiple columns / conditions) - OR operator

```python
filter_df = (df['country']=='Uzbekistan') | (df['region']=='Central')
df[filter_df]
```

## 8. Pivot Table

```python
df['sales'] = df['sales'].str.replace(',', '')
df['sales'] = df['sales'].astype(int)

df.pivot_table(values='profit',
               index='region',
               columns=df['order_date'].dt.to_period('Y'),
               aggfunc='mean',
               fill_value=0
)
```

## 9. Scatter Plot

```python
plt.scatter(df.sales, df.profit, s=45, color='r', edgecolors='purple', label="Scatter Chart")
plt.xlabel("Sales")
plt.ylabel("Profit")
# plt.grid(alpha=0.2)
plt.legend(loc="upper left")
plt.show()
```

## 10. Concatenate 2 columns

```python
df['con_col'] = df['country'] + '(' + df['region'] + ')'
df
```
